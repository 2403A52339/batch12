```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
n = 200
df = pd.DataFrame({
    'study_hours': np.random.normal(5, 2, n).clip(0),
    'sleep_hours': np.random.normal(7, 1.5, n).clip(0),
    'exam_score': np.random.normal(70, 10, n),
    'gender': np.random.choice(['Male', 'Female'], n)
})

df['exam_score'] = df['study_hours'] * 5 + np.random.normal(50, 10, n)

sns.pairplot(df, hue='gender', diag_kind='kde', palette='Set2')
plt.suptitle("Pairplot: Study Hours, Sleep Hours, Exam Score", y=1.02)
plt.show()
sns.lmplot(x='study_hours', y='exam_score', hue='gender', data=df,
           markers=['o', 's'], palette='Set1', height=6, aspect=1.2)
plt.title("Regression: Study Hours vs Exam Score by Gender")
plt.show()

sns.scatterplot(x='study_hours', y='exam_score', hue='gender', style='gender', data=df)
sns.regplot(x='study_hours', y='exam_score', data=df, scatter=False, color='black')
plt.title("Scatter + Regression Line: Study Hours vs Exam Score")
plt.show()
```
