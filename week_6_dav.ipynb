```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
blood_sugar = np.concatenate([
    np.random.normal(90, 10, 150),
    np.random.normal(140, 15, 50)
])

sns.kdeplot(blood_sugar, shade=True, color='crimson')
plt.title("KDE Plot of Blood Sugar Levels")
plt.xlabel("Blood Sugar (mg/dL)")
plt.ylabel("Density")
plt.show()
```
